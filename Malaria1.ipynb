# Why `Albumentation` instead of `Augmentation`
- Albumentation is the augmentation library widely used for image segmentation
- We used augmentation instead of albumentation because sometimes for image segmentation we resizd or rescale or orientation of bounding boxes distorted that's we now used albumentation as for it automatically adjust the boxes


```python

transforms = A.Compose(
    [
      A.Resize(global_image_size, global_image_size),

      A.OneOf([A.HorizontalFlip(),
                A.VerticalFlip(),], p = 0.3),
      
      A.RandomRotate90(),   
    #   A.RandomGridShuffle(grid=(3, 3), always_apply=False, p=0.5),
    #   A.RandomBrightnessContrast(brightness_limit=0.1,
    #                             contrast_limit=0.1,
    #                             always_apply=False, p=0.2),
    #   A.Cutout(num_holes=8, max_h_size=8, max_w_size=8, fill_value=0, always_apply=False, p=0.5),
      A.Sharpen(alpha=(0.2, 0.2), lightness=(0.2, 0.2), always_apply=False, p=0.2),
])

# ////////////////////////////////////////////////////////////////////////////////////////////////////////
# Taken from the albumentation documentation
def aug_fn(image):
    data = {"image":image}
    aug_data = transforms(**data)
    aug_img = aug_data["image"]
    aug_img = tf.cast(aug_img, tf.float32)
    return aug_img

# Taken from the albumentation documentation
def process_data_albumentation(image , label):
    aug_img = tf.numpy_function(func=aug_fn, inp=[image], Tout=tf.float32)
    return aug_img, label

# ////////////////////////////////////////////////////////////////////////////////////////////////////////

test_dataset_albu = (
    test_dataset
    .shuffle(buffer_size = 32)
    .map(process_data_albumentation)
    .prefetch(tf.data.AUTOTUNE)

)

# ////////////////////////////////////////////////////////////////////////////////////////////////////////
# This Plot function is for unbatached data
print("This is the dataset of the train set")
plt.figure(1)
for i , (image , label) in enumerate(test_dataset):
    if i<16:
        img_shape = image[1].shape
        plt.subplot(4 , 4 , i+1)    
        plt.imshow(image.numpy().astype("uint32"))
        i+=1
    else: break
print(img_shape)


print("This is the dataset of the test set Albu")
plt.figure(2)
for i , (image , label) in enumerate(test_dataset_albu):
    if i<16:
        img_shape = image[1].shape
        plt.subplot(4 , 4 , i+1)    
        plt.imshow(image.numpy().astype("uint32"))
        i+=1
    else: break
print(img_shape)

# ////////////////////////////////////////////////////////////////////////////////////////////////////////
```

## @tf.function    `Only used above tensorflow functions `
### `Only use tf.function above specific functions`
### `These function should only be using the tensoflow functioins`
### `Like resize transforms and others etc`
### `To print multiple times even with using tf.function we used specifal print function tf.print(" ") `
- This is used to make tensorflow function in the form of graph data structures. 
- We don't have to run same function again and again

```python
#By this you can see that the it is running only once

# @tf.function
# Only use tf.function above specific functions
# These function should only be using the tensoflow functioins
# Like resize transforms and others etc


@tf.function
def sample_function():
    print("I was here, This only print once")
    tf.print("Multiple Prints")


# You can see this function is printted once, because we are only concern with the return, the gra[h of function under tf.function is already stored in tensorflow function]

sample_function()
sample_function()
sample_function()
```

```python
class FeatureExtractor(Layer):
  def __init__(self, filters, kernel_size, strides, padding, activation, pool_size,):
    super(FeatureExtractor, self).__init__()

    drop_rate = 0.3
    self.dropout = Dropout(rate = drop_rate )  # Seed = None so that same layer dropout at the same rate 
    self.conv_1 = Conv2D(filters = filters, kernel_size = kernel_size, strides = strides, padding = padding, activation = activation)
    self.batch_1 = BatchNormalization()
    self.pool_1 = MaxPool2D (pool_size = pool_size, strides= 2*strides)
    self.conv_2 = Conv2D(filters = filters*2, kernel_size = kernel_size, strides = strides, padding = padding, activation = activation , kernel_regularizer = L2(0.01))
    self.batch_2 = BatchNormalization()
    self.pool_2 = MaxPool2D (pool_size = pool_size, strides= 2*strides)

  def call(self, x, training):

    x = self.conv_1(x)
    x = self.batch_1(x)
    x = self.pool_1(x)
    x = self.dropout(x)   
    x = self.conv_2(x)
    x = self.batch_2(x)
    x = self.pool_2(x)

    return x
  
class LenetModel_api(Model):
  
  def __init__(self ):
    super(LenetModel_api, self).__init__()
    self.feature_extractor = FeatureExtractor(8, 3, 1, "valid", "relu", 2)

    self.flatten = Flatten()
    self.dense_1 = Dense(100, activation = "relu" , kernel_regularizer = L2(0.01))
    self.batch_1 = BatchNormalization()

    self.dense_2 = Dense(10, activation = "relu")
    self.batch_2 = BatchNormalization()

    self.dense_3 = Dense(1, activation = "sigmoid")
    
  def call(self, x, training):

    x = self.feature_extractor(x)
    x = self.flatten(x)
    x = self.dense_1(x)
    x = self.batch_1(x)
    x = self.dense_2(x)
    x = self.batch_2(x)
    x = self.dense_3(x)

    return x
```

# Custom Loss and Accuracy Metrics
### For Custom Loss, we have to make out own custom Metrics. As `Metric` is interlinked with the loss during compiling
### As Metrics ``Accuracy`` is internlinked with the `Loss` 

```python
class Custom_bce(tf.keras.losses.Loss):
    def __init__(self):
        super(Custom_bce , self).__init__()    #This super runs the super class contructor first
    def call(self , y_true , y_pred):
        bce = tf.keras.losses.BinaryCrossentropy()
        return bce(y_true , y_pred)
    
# This is the custom accuracy function used the build in function
class custom_accuracy_one(tf.keras.metrics.Metric):
    def __init__(self , name = "custom_accuracy_one"):
        super(custom_accuracy_one , self).__init__()
        self.accuracy = tf.keras.metrics.BinaryAccuracy()
    def update_state(self , y_true , y_pred , sample_weight = None):
        self.accuracy.update_state(y_true , y_pred , sample_weight)
    def result(self):
        return self.accuracy.result()
    def reset_states(self):
        self.accuracy.reset_states()



# This is the custom accuracy function number 2
class custom_accuracy_two(tf.keras.metrics.Metric):
    def __init__(self , name = "custom_accuracytwo"):
        super(custom_accuracy_two, self).__init__()
        self.accuracy = self.add_weight(name = name , initializer = "zeros")

    def update_state(self , y_true , y_pred , sample_weight = None):
        output = binary_accuracy( tf.cast(y_true , dtype=tf.float32)  , tf.cast(y_pred , dtype=tf.float32))
        # print(output)
        # print("Testing--------")
        output_result =    tf.cast(tf.math.count_nonzero(output) , dtype= tf.float32)/ tf.cast(len(output) , dtype=tf.float32) 
        # print("Testing--------")
        self.accuracy.assign(tf.cast(output_result , dtype = tf.float32))
        # print("Testing--------")
        # print(output)
    def result(self):
        return self.accuracy
    def reset_states(self):
        self.accuracy.assign(0.)
    

loss = Custom_bce()
accuracy = custom_accuracy_two()
```

```python
def model_wandb(config):
  IM_SIZE = 128
  lenet_model = tf.keras.Sequential([
    InputLayer(input_shape = (IM_SIZE, IM_SIZE, 3)),

    Conv2D(filters = 6, kernel_size = 3, strides=1, padding='valid',
          activation = 'relu',kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),
    MaxPool2D (pool_size = 2, strides= 2),
    Dropout(rate = config["DROPOUT_RATE"]),

    Conv2D(filters = 16, kernel_size = 3, strides=1, padding='valid',
          activation = 'relu', kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),
    MaxPool2D (pool_size = 2, strides= 2),

    Flatten(),
    
    Dense( config["N_DENSE_1"], activation = "relu", kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),
    Dropout(rate = config["DROPOUT_RATE"]),
    
    Dense(config["N_DENSE_2"], activation = "relu", kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),

    Dense(1, activation = "sigmoid"),
  ])

  return lenet_model
     
```

```python
class LogImageCallback_wandb(Callback):

    def on_epoch_end(self, epoch, logs):
        # Assuming 'test_dataset' contains the test data and 'true_labels' contains the true labels
        # and 'functional_model' is the trained model used for prediction

        # Predict labels using the functional_model
        pred = model.predict(test_dataset)
        predicted_labels = np.round(pred).flatten()
        true_labels = np.concatenate([labels.numpy() for _, labels in test_dataset], axis=0)

        # Create the roc curve-----------------------------------
        fp , tp , thresholds =roc_curve(true_labels , predicted_labels) 
        plt.plot(fp, tp)
        plt.grid()
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")

        skip = 200
        for i in range(0, len(thresholds) , skip):
            plt.text(fp[i] , tp[i] , thresholds[i])

        buffer = io.BytesIO()
        plt.savefig(buffer, format = 'png')
    
        image_array = tf.image.decode_png(buffer.getvalue(), channels=3)
    
        images = wandb.Image(image_array, caption="ROC for epoch: {}".format(epoch))
        # clear the figure for the next plot
        plt.clf()
        threshold  =  0.5
        
        #Creating Custom Confusion Matrix ----------------------------
        cm = confusion_matrix(true_labels, predicted_labels > threshold)
        plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
        plt.title("Confusion Matrix")
        plt.colorbar()
        plt.xlabel("Predicted Labels")
        plt.ylabel("True Labels")
        buffer_confusion = io.BytesIO()
        plt.savefig(buffer_confusion, format = 'png')
    
        image_array_confusion = tf.image.decode_png(buffer_confusion.getvalue(), channels=3)
    
        images_confusion = wandb.Image(image_array_confusion, caption="Confusion Matrix for the  Epoch: {}".format(epoch))
   
        wandb.log({"conf_mat" : wandb.plot.confusion_matrix(probs=None,
                                                            y_true=true_labels,
                                                            preds=predicted_labels,
                                                            class_names=["Parasitized", "Uninfected"],
                                                            title="Confusion Matrix")}
                                                            )
        wandb.log({"roc" : images})
        wandb.log({"Confusion_Matrix" : images_confusion})
        
        
```

## Model Tuning using Wandb
- This global `CONFIGURATION` must be wandb configuration, so sweep can change its parameters in the main function 

```python

wandb.config = {
  "LEARNING_RATE": 0.001,
  "N_EPOCHS": 5,
  "BATCH_SIZE": 128,
  "DROPOUT_RATE": 0.0,
  "IM_SIZE": 224,
  "REGULARIZATION_RATE": 0.0,
  "N_FILTERS": 6,
  "KERNEL_SIZE": 3,
  "N_STRIDES": 1,
  "POOL_SIZE": 2,
  "N_DENSE_1": 100,
  "N_DENSE_2": 10,
}
CONFIGURATION = wandb.config


wandb_metrics = [
    TrueNegatives(name = "tn"),
    FalseNegatives(name = "fn"),
    TruePositives(name = "tp"),
    FalsePositives(name = "fp"),
    BinaryAccuracy(name = "accuracy"),
    Recall(name = 'recall'),
    Precision(name = 'precision'),
    AUC(name = 'auc')
]

sweep_configuration = {
    'method': 'random',
    'name': 'Malaria Detection Sweep 1',
    'metric': {'goal': 'maximize', 'name': 'accuracy'},
    'parameters': 
    {
        'REGULARIZATION_RATE': {'values': [0.01,0.001,0.0001]},
        'N_DENSE_1': {'values': [16,32,64]},
        'N_DENSE_2': {'values': [16,32,64]},
        'DROPOUT_RATE': {'min': 0.1 ,'max': 0.4 },
        'LEARNING_RATE': {'min': 1e-4 ,'max': 1e-2},
     
     }
}
```

```python
def model_wandb_tune(config):
  IM_SIZE = 128
  lenet_model = tf.keras.Sequential([
    InputLayer(input_shape = (IM_SIZE, IM_SIZE, 3)),

    Conv2D(filters = 6, kernel_size = 3, strides=1, padding='valid',
          activation = 'relu',kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),
    MaxPool2D (pool_size = 2, strides= 2),
    Dropout(rate = config["DROPOUT_RATE"]),

    Conv2D(filters = 16, kernel_size = 3, strides=1, padding='valid',
          activation = 'relu', kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),
    MaxPool2D (pool_size = 2, strides= 2),

    Flatten(),
    
    Dense( config["N_DENSE_1"], activation = "relu", kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),
    Dropout(rate = config["DROPOUT_RATE"]),
    
    Dense(config["N_DENSE_2"], activation = "relu", kernel_regularizer = L2(config["REGULARIZATION_RATE"])),
    BatchNormalization(),

    Dense(1, activation = "sigmoid"),
  ])

  return lenet_model
     
```

```python
def wandb_train():
    with wandb.init(project='Malaria Detection') as run:
        # This global configurartion must be wandb configuration
        CONFIGURATION = wandb.config # To set the global configuration, this global configurartion must be wandb configuration, these key mame should be equal to ones in the sweep configuration
        model = model_wandb_tune(CONFIGURATION )
        model.compile(
                optimizer= Adam(learning_rate = CONFIGURATION["LEARNING_RATE"]),
                loss='binary_crossentropy',
                metrics=['accuracy'],
            )
        model.summary()
        model.fit(train_dataset , epochs = 5 , validation_data = validation_dataset , callbacks = [WandbCallback()] , verbose = 1)

count = 5  
sweep_id = wandb.sweep(sweep=sweep_configuration)
wandb.agent(sweep_id , function = wandb_train , count = count)
wandb.finish()
    
```
